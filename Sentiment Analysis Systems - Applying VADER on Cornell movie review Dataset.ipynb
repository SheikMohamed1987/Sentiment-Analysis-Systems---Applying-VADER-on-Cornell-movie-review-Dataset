```python
import nltk
```

```python
from nltk.sentiment import vader
```

```python
sia = vader.SentimentIntensityAnalyzer()
def vaderSentiment(review):  
    return sia.polarity_scores(review)['compound']
```

```python
vaderSentiment("This is my first Excellent score")
```

```python
negativeReviewsFileName =  'C:/Users/Sheik/Documents/DataSets/rt-polaritydata/rt-polaritydata/rt-polarity.neg'
```

```python
with open(negativeReviewsFileName,'r') as f:     negativeReviews = f.readlines()
```

# Apply VADER to reviews

```python
def getReviewSentiments(sentimentCalculator):   
    negReviewResult =    [sentimentCalculator(oneNegativeReview) for oneNegativeReview in negativeReviews]   
    posReviewResult =    [sentimentCalculator(onePositiveReview) for onePositiveReview in positiveReviews]   
    return {'results-on-positive':posReviewResult, 'results-on-negative':negReviewResult}
```

```python
negReviewResult =    [vaderSentiment(oneNegativeReview) for oneNegativeReview in negativeReviews]   
posReviewResult =    [vaderSentiment(onePositiveReview) for onePositiveReview in positiveReviews] 
```

```python
getReviewSentiments(vaderSentiment)
```

```python
pctTruePositive = float(sum(x > 0 for x in posReviewResult))/len(posReviewResult)
pctTruePositive
```

```python
pctTrueNegative  = float(sum(x < 0 for x in negReviewResult))/len(negReviewResult)
pctTrueNegative 
```

```python
totalAccurate =  float(sum(x > 0 for x in posReviewResult)) + float(sum(x < 0 for x in negReviewResult)) 
total = len(posReviewResult) + len(negReviewResult) 
print ("Overall accuracy=" +  "%.2f" % (totalAccurate*100/total) + "%")
```
